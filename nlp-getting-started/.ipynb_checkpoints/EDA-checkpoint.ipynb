# Text

The 'text' column contains the individual tweets which we need to classify. This is likely to be the most important column for classification so we need to understand what data cleaning needs to be done and which additional features can be extracted to help the model.

## Meta Features

There are a number of features which can be extracted from a tweet which may be helpful in correctly identifying genuine disaster tweets. For example, disaster tweets could be written in a more formal way (e.g. from a News agency), could have more characters to describe the event in more detail etc. 'Spammy' tweets on the other hand may have erratic punctuation, typos, repetition of words etc.

The following simple features will be generated:  
- Counts
    - `word_count` number of words in the tweet  
    - `unique_word_count` number of unique words in the tweet  

- Reading Difficulty
    - `readability_*` what level of education would be able to understand the tweet



**Counts**

```python
#word count
train_df['counts_word'] = train_df['text'].apply(lambda x: len(str(x).split()))

#unique_word_count
train_df['counts_unique_word'] = train_df['text'].apply(lambda x: len(set(x.split())))

```

```python
count_cols = [col for col in train_df.columns if 'counts' in col]

plot_column_distributions(count_cols,train_df,figsize=(10,5))
```

Word count and unique words seems to only exhibit small differences between the two categories.

**Readability Metrics**

Will test all of the inbuilt functions in the textstat library

See [textstat](https://pypi.org/project/textstat/) library for metric definitions

```python
import textstat

metrics = [textstat.flesch_reading_ease,
              textstat.smog_index,
              textstat.flesch_kincaid_grade,
              textstat.coleman_liau_index,
              textstat.automated_readability_index,
              textstat.dale_chall_readability_score,
              textstat.difficult_words,
              textstat.linsear_write_formula,
              textstat.gunning_fog,
              textstat.text_standard]

for metric in metrics:
    train_df[f"readability_{metric.__name__}"] = train_df['text'].apply(metric)
```

```python
readability_cols = [col for col in train_df.columns if 'readability_' in col and 'text_standard' not in col]
plot_column_distributions(readability_cols,train_df,figsize=(15,20))
```

There definitely seems to be a difference between the readability of the tweet and the whether or not the tweet is a legitemate disaster tweet. On every metric, disaster tweets have a more difficult set of words. This is probably because many diaster tweets are formal (e.g. from news agencies or from concerned members of the public reporting on what they have seen). 

This looks like it could be a useful feature in the model, however, it is very likely that these features are highly correlated with each other. So we may only need one feature to capture this information.

```python
plt.figure(figsize=(12,8))

corr = train_df[readability_cols].corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

sns.heatmap(corr,cmap='Blues',annot=True,fmt='.2f',mask=mask)
plt.title("Correlation between readability metrics",pad=20,fontsize=18)
plt.show()
```

As expected, readability metrics are highly correlated with each other. The 'flesch_reading_ease' metric is inversely correlated with the other features, however, this is expected as it is the only metric measuring how *easy* the tweet is to read, whereas all the others are measuring *difficulty*. The smong_index metric is not very correlated to the other features, however, looking at the distributions above, it looks like this feature is not very different between genuine and non-genuine tweets and is unlikely to be useful.

Of the readability features, the flesch_kincaid_grade feature has the highest correlations with all other features so this can be used as a good proxy for all the other readability features
